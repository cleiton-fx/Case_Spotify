```python
# Importando bibliotecas necessárias

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregando dados

dados = pd.read_csv('tracks.csv')

dados.head()
```

```python
#Checando valores nulos

dados.isnull().sum()
```

```python
#Checando tipo de dados e informações 

dados.info()
```

```python
#Checando os ultimos dados por popularidade

sorted_df = dados.sort_values('popularity', ascending = True).head()

sorted_df
```

```python
dados.describe().transpose()
```

```python
#Checamos aqui as músicas mais populares acima de  90

mais_popular = dados.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)

mais_popular[:10]
```

```python
#Checando tipos de dados

dados.dtypes
```

```python
#Transformando a coluna release_date em índice e mudando seu formato para datetime 


dados.set_index('release_date', inplace = True)
dados.index = pd.to_datetime(dados.index)
dados.head()
```

```python
#Localizando determinado artista, neste caso o  artista 18

dados[['artists']].iloc[18]
```

```python
#renomeando a duração, colocando um (ms) para mostrar que esta em milisegundos
#Aplicando formula com lamba para arredondar valores

dados['duration'] = dados['duration_ms'].apply(lambda x: round(x/1000))
dados.drop('duration_ms', inplace = True, axis = 1)
```

```python
#Observando primeiras linhas de durção

dados.duration.head()
```

```python
#Removendo variáveis nao necessárias
#verificando as correlações entre as seguintes variáveis
#https://seaborn.pydata.org/tutorial/color_palettes.html

corr_df = dados.drop(["key", "mode", "explicit"], axis = 1).corr(method = 'pearson')

plt.figure(figsize = (15,8))
heatmap = sns.heatmap(corr_df,annot = True, fmt = '.1g', vmin = -1, vmax = 1, center = 0, cmap = 'inferno', linewidths = 1, linecolor ='black')
heatmap.set_title("Correlação entre Variáveis")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

```python
#Ajuste para  de regressão

dados_a = dados.sample(int(0.004*len(dados)))
```

```python
print(len(dados_a))
```

```python
# Observando variáveis que obtiveram maior correlação
# O gráfico mostra claramente um regressão linear, assim confirmando os dados pelo heatmap

plt.figure(figsize = (10,6))
sns.regplot(data = dados_a, y = 'loudness', x = 'energy', color = 'c').set(title = 'Correlação entre Loudness e  Energy')

```

```python
#Observando um exemplo onde nao há correlação entre as variáveis

plt.figure(figsize = (10,6))
sns.regplot(data = dados_a, y = 'popularity', x = 'acousticness', color = 'b').set(title = 'Correlação entre Popularity e  Acousticness')
```

```python
#Pegando as datas 
#Mudando formato das datas para datetime
#Plotando numero de musicas conforme suas datas

dados['dates']=dados.index.get_level_values('release_date')
dados.dates =pd.to_datetime(dados.dates)
years= dados.dates.dt.year


sns.displot(years, discrete = True, aspect = 2, height = 5,kind = 'hist').set(title = 'Número de músicas por ano')
```

```python
# A intenção dete gráfico será gerar um array contendo valores de Ano x Duração contendo os dados totais de duração das músicas 

total_dr = dados.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y= total_dr, ax = ax, errwidth = False).set(title = 'Ano x Duração')
plt.xticks(rotation = 60)
```

```python
# Agora vamos carregar o  DB onde esta inserido o genero e outras variáveis
# São pouco mais de 232k
#https://www.kaggle.com/datasets/zaheenhamidani/ultimate-spotify-tracks-db
# Aqui podemos analiser por gênero



df_genero = pd.read_csv('feature.csv')

df_genero.head()
```

```python
#Vamos visualizar graficamente

plt.title('Duração das  músicas em diferentes gêneros')
sns.color_palette('flare', as_cmap = True)
sns.barplot(y = 'genre', x = 'duration_ms', data = df_genero)
plt.xlabel('Duração em Milissegundos')
plt.ylabel('Gêneros')
```

```python
# E se quisermos visualisar os top 5 gêneros por popularidade?
# Tambem podemos visualisar isso graficamente

sns.set_style(style ='darkgrid')
plt.figure(figsize = (10,5))
famoso = df_genero.sort_values('popularity', ascending = False).head(10)
sns.barplot(y= 'genre', x ='popularity', data = famoso).set(title = 'Top 5 Gêneros por popularidade')
```
